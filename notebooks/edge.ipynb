## Load data:

```python
import os
from PIL import Image

def load_images(image_dir):
    images = []
    for file in os.listdir(image_dir):
        if file.endswith(".jpg") or file.endswith(".png"):
            path = os.path.join(image_dir, file)
            img = Image.open(path)
            images.append(img)
    return images

load_images('../data/raw/image/frames')
```

## Data quality:

```python
def check_images(images, labels):
    print("Number of images:", len(images))
    print("Number of labels:", len(labels))

    # Consistency
    if len(images) != len(labels):
        print("WARNING: mismatch between images and labels")
```

## Preprocessing:

```python
def preprocess_images(images, size=(320, 240)):
    processed = []
    for img in images:
        img = img.resize(size)
        img = img.convert("RGB")
        processed.append(img)
    return processed

preprocess_images(load_images('../data/raw/image/frames'))
```
